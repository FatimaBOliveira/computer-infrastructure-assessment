**Task 3: Formatting Timestamps**

Run the date command again, but this time format the output using YYYYmmdd_HHMMSS (e.g., 20261114_130003 for 1:00:03 PM on November 14, 2026). Refer to the date man page (using man date) for more formatting options. (Press q to exit the man page). Append the formatted output to a file named formatted.txt.

Steps:

- now to append the output to the new file, I write:
```bash

```
> date +:"..." display date and/or time in the format defined by the user.

> %Y: print year.

> %m: print month.

> %d: print day.

> %H: print hour.

> %M: print minutes.

> %S: print seconds.

### Analysing data
---
Python's library, pandas, can load the data with [pd.read_json](https://pandas.pydata.org/docs/reference/api/pandas.read_json.html). To identify the file, the path and the name file are needed. After that, it's possible to make some analysis.

````python
# Library.
import pandas as pd

# File to use.
weatherfile="./data/weather/20241101_164019.json"
````

````python
# load file.
df=pd.read_json(weatherfile)

# Check.
df.head(3)
````

In this DataFrame, there's a lot of weather information: name location, temperature, symbol, weather description and text, wind speed, wind gust, cardinal wind direction, wind direction, humidity, rainfall, pressure, day name, date and report time. 

It's also possible to check the data type with [.dtypes](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dtypes.html).

````python
# Inspect the types.
df.dtypes
````

The types of variables present in this data are: object, int, and datetime.
It's possible to make a [statistical description](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html) of variables when these are ints or floats, as demonstrated below.

````python
# Describe the data.
df.describe()
````

When doing the describe function in the Data Frame, Pandas generates some statistics, like the count, mean, min, max, standard deviation and the percentile 25%, 50% and 75% of the values given. [W3schools](https://www.w3schools.com/python/python_ml_percentile.asp) explains that "Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than."
The only variable that we can ignore it's the date because it doesn't produce any relevant information for this type of statistic.

The data can also be analysed visually through a plot with the help of Matplotlib. Below, two variables are represented, time and humidity. To make the plot easier to read, I added a title, axes labels, [x ticks rotation](https://www.geeksforgeeks.org/how-to-rotate-x-axis-tick-label-text-in-matplotlib/), a legend and a [grid](https://www.w3schools.com/python/matplotlib_grid.asp).

````python
# Import library.
import matplotlib.pyplot as plt

# Define variables to plot
plt.plot(df["reportTime"], df["humidity"], label="Humidity")

# Add a title.
plt.title("Humidity over time")

# Add axes labels.
plt.xlabel("Time")
plt.ylabel("Humidity")

# X tick orientation.
plt.xticks(rotation=75)

# Add a legend.
plt.legend()

# Add grid.
plt.grid(which="both", color="grey", linestyle='--', linewidth=0.5)

# Plot.
plt.show()
````

As demonstrated, the highest humidity was registered between 7H and 9H, and the lowest at 15H.

##  Project - workflow
---
In this part, I will explain how I created a workflow and how it's running.
First of all, it was necessary to create a new directory, .githib/workflows. This can be created through the command line or in the explorer section, in both GitHub Codespaces or Visual Studio Code. After that, I created a file with the configuration language necessary, weather-data.yml, containing the instructions to be deployed. These can be automated with GitHub Actions, which provides a virtual machine available all the time to execute the jobs scheduled on the YAML file. 

While writing my weather-data.yml file, I followed the instructions given by the Professor, searched for some information and asked ChatGPT to create an example. Below I will explain each section of this file:

1.
```YAML
name: Run weather.sh script, commit and push changes
```
This part shows the name of the workflow, used to identify it in GitHub Actions.

https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#name

2.
```YAML
permissions: 
  contents: write
```
GitHub Actions will need "permissions" activated, as this indicates that the user permits them to do the jobs. The "contents" informs them that they can manipulate the contents in the repository, and "write" indicates that these files can be committed and pushed.

https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#permissions

3.
```YAML
on:
  schedule:
    - cron: '0 10 * * *'
  workflow_dispatch:
```
The "on" will show the conditions that will initiate the workflow.
 
Then, "schedule" will set the time that the workflow will run, and "cron" will specify the time in the following order: minute, hour, day of the month, month, and day of the week.

After that, "workflow_dispatch" is necessary to enable the user to trigger the workflow manually in the GitHub Actions.

https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#on
https://docs.github.com/en/actions/writing-workflows/choosing-when-your-workflow-runs/events-that-trigger-workflows#schedule
https://en.wikipedia.org/wiki/Cron
https://crontab.guru/
https://docs.github.com/en/actions/writing-workflows/choosing-when-your-workflow-runs/events-that-trigger-workflows#workflow_dispatch
https://docs.github.com/en/actions/managing-workflow-runs-and-deployments/managing-workflow-runs/manually-running-a-workflow#running-a-workflow

4.
```YAML
jobs:
  run-weather-script:
    runs-on: ubuntu-latest
```
In this section, the "jobs" will be shown. "run-weather-script" is the name of the job. The part "runs-on" will tell which type of operating system to use to execute this job, "ubuntu-latest" will indicate to use the latest version of Ubuntu, as this is one of the most used ones.

https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#jobs
https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#jobsjob_id

https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#standard-github-hosted-runners-for-public-repositories

5.
```YAML
    steps:
    - name: Checkout repository
      uses: actions/checkout@v4
```
The job needs to have the "steps" identified that will tell the machine what to do, as GitHub states: "Steps can run commands, run setup tasks, or run an action in your repository, a public repository, or an action published in a Docker registry."

In each step, a "name" needs to be attributed, then "uses" will make an action run, and "actions/checkout@v4" informs that these actions will check out my repository with the latest version of the action, that is 4.

https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#jobsjob_idsteps
https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#jobsjob_idstepsname

https://github.com/actions/checkout?tab=readme-ov-file

6.
```YAML
    - name: Run weather.sh script
      run: |
        chmod +x weather.sh
        ./weather.sh
```
To create another step, another "name" needs to be applied to that step. The code "run" will trigger a command line, "|" shows that there's more than one command to run. "chmod +x weather.sh" will make the script executable to the computer and "./weather.sh" shows the location of this file.


https://docs.github.com/en/actions/writing-workflows/workflow-syntax-for-github-actions#jobsjob_idstepsrun

7.
```YAML
    - name: Commit and Push changes
      run: |
        git config --global user.name 'github-actions[bot]'
        git config --global user.email '41898282+github-actions[bot]@users.noreply.github.com'
        git add .
        git commit -m "Run weather.sh script and commit changes"
        git push
      env:
        GITHUB_TOKEN: ${{ secrets.GITHUB_TOKEN }}
```
Another step was added with more command lines, "git config --global user.name 'github-actions[bot]'" and "git config --global user.email '41898282+github-actions[bot]@users.noreply.github.com'" allows the machine to use git, then "git add ." will add files into my repository. Finally, "git commit -m "Run weather.sh script and commit changes" will commit the file with that description, and in the end, it will push it with "git push"

For this step, "env" indicates the variable used and "GITHUB_TOKEN: ${{ secrets.GITHUB_TOKEN }}" enables GitHub to create an access token for GitHub Actions, that will install GitHub app in the repository and will do the actions necessary to run the workflow. All this process is done in secret because we can't see it visually.
